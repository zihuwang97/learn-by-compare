```python
"""
Plot LbC model latent representations using UMAP
"""

import torch
import umap
import umap.plot
import numpy as np
from models import encoder, predictor
import bokeh

device = 'cuda:1'

# load encoder
enc = encoder(input_feature_num=520, hdn_size=80)
state_dict = torch.load('train_log/24081923:36:13/checkpoint_1999.pth.tar' , map_location="cpu")
enc.load_state_dict(state_dict)
enc.to(device)
# predictor = predictor(embed_size=32, target_size=1)
# predictor_state = torch.load('best_predictor.pth.tar', map_location="cpu")
# predictor.load_state_dict(predictor_state)
# predictor.to(device)

# load data
data = np.load('BFMC-1000-POST-LAY/inputdata.npy')
label = np.load('BFMC-1000-POST-LAY/outputdata.npy')
data= torch.as_tensor(data, dtype=torch.float).to(device)
label = torch.as_tensor(label, dtype=torch.float)

# rep = enc(data) 
# pred = predictor(rep)
# error = torch.abs(pred.reshape(-1) - label.to(device))
# eligible_idx = error<1.1
# print(torch.sum(eligible_idx))

# get embeddings
embedding = enc(data).detach().to('cpu')
mapper = umap.UMAP(random_state=42, min_dist=5e-3, spread=10).fit(embedding)
umap.plot.output_notebook()
p = umap.plot.interactive(mapper, labels = label[:,1].reshape(-1), point_size=20)
bokeh.plotting.show(p)
# umap.plot.show(p)
```

```python
"""
Plot END-TO-END model latent representations using UMAP
"""

import torch
import umap
import umap.plot
import numpy as np
from models import whole_model

device = 'cuda:1'

# load encoder
model = whole_model(520, 80, 2)
state_dict = torch.load('train_log/l1_checkpoints/checkpoint_9999.pth.tar' , map_location="cpu")
model.load_state_dict(state_dict)
enc = model.enc
enc.to(device)

# load data
data = np.load('BFMC-1000-POST-LAY/inputdata.npy')
label = np.load('BFMC-1000-POST-LAY/outputdata.npy')
data= torch.as_tensor(data, dtype=torch.float).to(device)
label = torch.as_tensor(label, dtype=torch.float)

# get embeddings
embedding = enc(data).detach().to('cpu')
mapper = umap.UMAP(random_state=42).fit(embedding)
# umap.plot.points(mapper, labels = label, show_legend=False)
umap.plot.output_notebook()
p = umap.plot.interactive(mapper, labels = label[:,0].reshape(-1), point_size=20)
bokeh.plotting.show(p)
```
